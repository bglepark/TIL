### 연습문제1.
- 두 개의 데이터프레임을 만들고 merge 명령을 병합한다.
- 단, 데이터 프레임은 다음 조건을 만족해야 한다.
    1. 각각 5X5 이상의 크기를 가진다.
    2. 공통열을 하나이상 갖는다. 
        - 다만 공통열의 이름은 서로 다르게 할 것
    3. merge의 경우를 inner, outer, left, right 4개의 형태로 출력할 것
    4. 지정된 인덱스와 컬럼명을 갖는다. 

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '대구'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '강릉']]

df1 = pd.DataFrame(data, 
                   index=['홍길동', '이몽룡', '성춘향','변학도','김연아'], 
                   columns=["나이", "몸무게", "키", "성별" ,"주소"])
df1
```

```python
data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3,180.1,  '남', '울산'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '광주']]
df2 = pd.DataFrame(data,
                   index=[1,2,3,4,5],
                   columns=["나이", "몸무게", "키", "성별" ,"지역"])
df2
```

```python
df1.merge(df2 , left_on='주소' , right_on='지역')
df1.merge(df2 , left_on='주소' , right_on='지역' , how='outer')
df1.merge(df2 , left_on='주소' , right_on='지역' , how='left')
df1.merge(df2 , left_on='주소' , right_on='지역' , how='right')
```

### 연습문제2. 
- 어느회사의 전반기(1월-6월)실적을 나타내는 데이터프레임과
후반기(7월-12월)실적을 나타내는 데이터 프레임을 작성한 후
합친다(단순병합)
- 실적 정보는 "매출","비용","이익"으로 이루어진다.
- 이익=매출-비용 : 가공필드로 생성
    
- 또한 1년간의 총 실적을 마지막 열로 덧붙인다.

**상반기(1-6월) 실적 데이터프레임 생성**
- month 변수 생성

```python
month = [str(i)+'월' for i in range(1,13)]
month
```

- 1~6월의 매출, 비용 데이터 생성(딕셔너리)

```python

# 매출, 비용 data random  생성
np.random.seed(3)
data = {}
for i in month[0:6]:
    data[i]=np.random.randint(1,100,2)

data
```

- 1~6월의 매출과 비용 데이터프레임 생성

```python
# 데이터프레임 생성
data= pd.DataFrame(data , index=['매출','비용'])
data
```

- '이익' 행이 추가된 데이터 프레임 작성

```python
# '이익' 행 생성 : 매출 - 비용
benefit = {}
for i in month[0:6]:
    benefit[i] = data[i][0]-data[i][1]
benefit = pd.DataFrame(benefit , index = ['이익'])
data = pd.concat([data,benefit])
data
```

**하반기(7~12월) 실적 데이터프레임 생성**
- 매출, 비용으로 구성

```python
# 매출, 비용 data random  생성


# 데이터프레임 생성


# 가공 행 이익 생성

np.random.seed(10)
data2 = {}
for i in month[6:]:
    data2[i]=np.random.randint(1,100,2)
data2
data2=pd.DataFrame(data2 , index=['매출','비용'])
data2

benefit2 = {}
for i in month[6:]:
    benefit2[i] = data2[i][0]-data2[i][1]
benefit2 = pd.DataFrame(benefit2 , index = ['이익'])
data2 = pd.concat([data2,benefit2])
data2

```

**상반기 실적 데이터프레임과 하반기 실적 데이터프레임 연결**

```python
result = pd.concat([data,data2] , axis=1)
result
```

- '총실적' 열 추가

```python
final1 = result.loc['매출'].sum()
final2 = result.loc['비용'].sum()
final3 = final1 - final2
result['총실적'] = [final1,final2,final3]
result
```

## 데이터 그룹의 변형 : transform()

### 데이터 변형의 일반적인 과정

: transform() 메소드는 데이터프레임의 모든 값에 함수를 적용하며, 다음의 특성을 갖는 새 데이터프레임을 반환한다

- 모든 그룹의 인덱스가 합쳐진 인덱스를 갖는다
- row의 개수는 모든 그룹의 row 개수의 총합과 같다
- 그룹화 대상이 아닌 컬럼도 함수가 성공적으로 적용된다면 결과에 포함되며, 그렇지 않은 컬럼은 삭제될 수 있다

**예제 데이터1**

```python
trans_data = pd.DataFrame({'Label' : ['A','C','B','A','C'],
                           'Values': [0, 1, 2, 3, 4],
                           'Values2' : [5, 6, 7, 8, 9],
                           'Other': ['foo','bar','baz','fiz','buz']},
                         index=list('VWXYZ'))
trans_data
```

**Label 컬럼으로 그룹화**

```python
Label = trans_data.groupby(['Label'])
Label
```

```python
def print_groups(group_object):
    for name , group in group_object:
        print(name)
        print(group[:5])

print_groups(Label)
```

```python
Label.groups
```

### 각 값에 10을 더하는 함수를 적용해 데이터프렘임 변형 수행

```python
trans_data[['Values','Values2']].transform(lambda x:x+10 , )
```

-> 'Label'과 'Other' 컬럼은 문자열 값들이므로 함수 적용에 실패하여 두 컬럼은 결과에서 제외됨

### 그룹의 평균으로 결측값 채우기

**예제 데이터2**

```python
trans_data2 = pd.DataFrame({'Label' : list('ABABAB'),
                            'Values' : [10, 20, 11, np.nan, 12, 22]})
trans_data2
```

```python
Label2 = trans_data2.groupby('Label')
print_groups(Label2)
```

**각 그룹의 평균 계산**

```python
Label2.mean()
```

**결측치(Nan) 채우기**

- 그룹별 평균으로 채우기

```python
fillna = Label2.transform(lambda x:x.fillna(x.mean()))
fillna
```

## 그룹 필터링 : filter()

- 데이터그룹을 선택적으로 삭제

**예제 데이터3**

```python
df3 = pd.DataFrame({'Label' : list('AABCCC'),
                    'Values' : [1, 2, 3, 4, np.nan, 8]})
df3
```

**아이템의 개수가 지정한 최소한의 수를 넘지 않는 그룹을 제외**

```python
df3.groupby('Label').filter(lambda x:x.Values.count() > 1)
```

**NaN이 하나라도 존재하는 그룹 제외**

```python
df3.groupby('Label').filter(lambda x:x.Values.isnull().sum() ==0)
```

**전체 데이터셋 평균과의 차이가 2.0이 넘는 그룹 평균을 가진 그룹만 필터링**

```python
df3.groupby('Label').mean()
```

```python
# 전체 평균
gmean = df3.groupby('Label').mean().mean()
gmean
```

```python
df3.groupby('Label').filter(lambda x: x.Values.mean()-gmean >2.0)
```
