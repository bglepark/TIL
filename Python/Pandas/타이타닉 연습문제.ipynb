### 연습문제 : 수치형-> 범주형 데이터 생성

: 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive), 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.

- bins=[1,15,25,35,60,99]
- labels=['미성년자','청년','중년','장년','노년']
- 각 그룹별 비유의 전체 합은 1이 되어야 한다.

```python
# 타이타닉 승객 dataset 읽어오기
import seaborn as sns 
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
live = titanic[titanic['alive']=='yes']
live
```

```python
dead = titanic[titanic['alive']=='no']
dead
```

```python
# 라벨과 범위 간격 설정
bins=[1,15,25,35,60,99]
labels = ['미성년자' , '청년', '중년' , '장년' ,'노년' ]

min(bins)
live['age'].min()
dead['age'].min()

max(bins)
live['age'].max()
dead['age'].max()

## bins 수정
bins=[0,15,25,35,60,99]
```

```python
cut_dead = pd.cut(list(dead['age']) , bins , labels=labels)
cut_live = pd.cut(list(live['age']) , bins , labels=labels)

cut_dead.categories
cut_live.categories
```

```python
dead['age_cut']=cut_dead
live['age_cut']=cut_live

dead.head()
live.head()
```

```python
## 사망자 그룹에 대한 비율 계산

dead['age_cut'].value_counts()
dead_total = dead['age_cut'].count().sum()
dead['age_cut'].value_counts()/dead_total
(dead['age_cut'].value_counts()/dead_total).sum()
```

```python
## 생존자 그룹에 대한 비율 계산

live['age_cut'].value_counts()
live_total = live['age_cut'].count().sum()
live['age_cut'].value_counts()/live_total
(live['age_cut'].value_counts()/live_total).sum()
```
