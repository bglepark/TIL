벡터화 인덱싱 : Boolean selection
- boolean Series 가 []와 함께 사용되면 True 값에 해당하는 값만 새로 반환하는 Series 객체에 포함됨
- 다중조건의 경우 , &(and) , |(or)를 사용하여 연결 가능

```python
# 벡터화 인덱싱
# s 시리즈 값 중 2500000 보다 크고 5000000 보다 작은 원소를 추출

condition = (s>2500000) & (s <5000000)
# condition = (s>25e5) & (s <50e5) 이렇게 표현도 가능!! 

# e 를 사용하자
```

```python
s[condition]
```

```python
# 바로 index값을 다양하게 사용할 수 있다
s0 = pd.Series(np.arange(10) , np.arange(10)+1)
s0
```

```python
# 5보다 큰가?
s0>5
```

```python
s0[s0>5]
```

```python
# 짝수 추출
s0[s0%2==0]
```

```python
# 인덱스가 5보다 큰가?
s0.index>5
```

```python
s0[s0.index>5]
```

```python
s0[(s0>5)&(s0<8)].reset_index(drop=True)
```

```python
# 값이 7이상인 수들의 합 구하기
s0[s0>=7].sum()
```

```python
(s0>=7).sum() # 값이 7이상인 수들의 갯수(True의 결과)
```

**날짜 생성 함수 : data_range()**
** 교재 p.242 참조
- 판다스 패키지의 date_range 함수(날짜생성)
- pd.date_range(start=None , end=None , periods=None , freq='D)
- freq 약어 옵션은 교재 참조

```python
# 일별 간격으로 날짜 생성
# dtype='datetime64[ns] 데이트 타임 형태로 생성'
# default => D로 지정되어있음

date = pd.date_range(start = '2021-10-01' , end='2021-10-20')
date
```

```python
type(date)
```

```python
# 하루 주기로 '2021-10-01' 에서 '2021-10-20' 사이의 일자 생성

pd.date_range(start='2021-10-01' , end = '2021-10-20' , freq='D')
```

```python
# 3일 주기로 '2021-10-01' 에서 '2021-10-20' 사이의 일자 생성

pd.date_range(start='2021-10-01' , end = '2021-10-20' , freq='3D')
```

- 주간격으로 날짜 지정
freq = 'W' : 일요일 기준으로 일주일 주기 생성
freq = 'W-MON' : 월요일 기준으로 일주일 주기로 생성

```python
# 2021-10-01을 기준으로 1주일씩 증가하는 4개의 1주 시작일(일요일)
# 2021-10-01 은 금요일

pd.date_range(start='2021-10-01' , end = '2021-10-20' , freq='W')
```

```python
pd.date_range(start='2021-10-01' , end = '2021-10-20' , freq='W-SUN')
```

```python
# 2021-10-01을 기준으로 1주일씩 증가하는 4개의 1주 시작일(월요일)
pd.date_range(start='2021-10-01' , end = '2021-10-20' , freq='W-MON')
```

```python
# 2021-10-01을 기준으로 1주일씩 증가하는 4개의 1주 시작일(금요일)
pd.date_range(start='2021-10-01' , end = '2021-10-20' , freq='W-FRI')
```

```python
# 2021-10-07을 기준으로 1주일씩 증가하는 4개의 1주 시작일(월요일)
pd.date_range(start='2021-10-07' , periods=4 , freq='W-MON')
```

**월간격으로 날짜 생성
- freq = 'M' : 월말 날짜 기준 주기
- freq='BM' : 업무 월말 날짜 기준 주기
- freq='MS' : 월초 날짜 기준 주기

```python
# 2021-10-07을 기준으로 한 달 간격의 일자 4개 생성 : 월말 기준

pd.date_range(start='2021-10-07' , periods=4 , freq='M')
```

```python
# 2021-10-07을 기준으로 한 달 간격의 일자 4개 생성 : 월초 기준

pd.date_range(start='2021-10-07' , periods=4 , freq='MS')
```

```python
# 2021-10-07을 기준으로 3 달 간격의 일자 4개 생성 : 월말 기준

pd.date_range(start='2021-10-07' , periods=4 , freq='3M')
```

분기 기준
freq = Q : 분기끝

freq = BQ : 업무 분기 끝 날짜 기준 주기

freq = QS : 분기 시작 날짜 기준 주기

분기 기준
freq = A : 연말 기준 주기

freq = AS : 연초 기준 주기

freq = BAS : 업무 일년 시작 날짜 기준

시간 기준
freq = H : 시간 주기

freq = BH : 업무 시간 주기

freq = T 또는 min : 분 주기

freq = S : 초 주기

```python
pd.date_range(start='2021-10-07 08:00' , periods=10 , freq='H')
```

```python
pd.date_range(start='2021-10-07 08:00' , periods=10 , freq='BH')
# 업무 시간은 9to5으로 되어있음
```

```python
pd.date_range(start='2021-10-07 08:00' , periods=10 , freq='30T')
# 30분 간격
```
