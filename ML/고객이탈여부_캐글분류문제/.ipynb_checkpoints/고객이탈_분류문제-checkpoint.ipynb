```python
# 라이브러리 import

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
```

```python
X_train = pd.read_csv('X_train.csv')
X_test = pd.read_csv('X_test.csv')
y_train = pd.read_csv('y_train.csv')
```

```python
X_train.head()
```

```python
# CutomerId는 최종 제출 파일에 필요한 정보이므로 삭제
X_train.drop(columns=['CustomerId'] , inplace=True)
y_train.drop(columns=['CustomerId'] , inplace=True)
Id = X_test['CustomerId']
X_test.drop(columns=['CustomerId'] , inplace=True)
```

```python
#훈련 데이터셋에서 범주형 데이터가 있는 컬럼의 데이터 구성 내용 및 오염 유무를 확인함
for col in X_train.select_dtypes(include=object).columns:
    target = X_train[col]
    print('***************************************')
    print(col)
    print(target.nunique())
    print(target.value_counts())

print(X_train.isna().sum())
print(X_train.info())

# Surname : 2289 개의 다른 데이터 -> 삭제
# Geography -> Labeling
# Gender -> 변수 이름 통일화 필요
# 수치형 변수에 결측치 없음
```

```python
# Surname 삭제
X_train.drop(columns=['Surname'] , inplace=True)
X_test.drop(columns=['Surname'] , inplace=True)
```

```python
# Gender 컬럼 변수 통일화 (Male , Female)

X_train['Gender'].replace({' male' : 'Male', 'female' : 'Female'} , inplace=True)
X_test['Gender'].replace({' male' : 'Male', 'female' : 'Female'} , inplace=True)
```

```python
for col in X_train.select_dtypes(include=object).columns:
    target = X_train[col]
    print('***************************************')
    print(col)

# 남아 있는 범주형 컬럼 : Geography , Gender
```

```python
for col in X_train.select_dtypes(include=['int64' , 'float64']).columns:
    target = X_train[col]
    print('***************************************')
    print(col)
    print(target.describe())
    print(target.std())

# std 값이 매우 큰 컬럼 : Balance , EstimatedSalary
```

```python
# std 값이 매우 큰 컬럼 : Balance , EstimatedSalary 에 대해 StandardScaler 적용

Scaler = StandardScaler()

X_train['Balance'] = Scaler.fit_transform(X_train[['Balance']])
X_train['EstimatedSalary'] = Scaler.fit_transform(X_train[['EstimatedSalary']])

X_test['Balance'] = Scaler.transform(X_test[['Balance']])
X_test['EstimatedSalary'] = Scaler.transform(X_test[['EstimatedSalary']])
```

```python
# 범주형 변수에 원핫 인코딩

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
# 훈련데이터 : 검증 = 7:3

X_train_model, X_train_val, y_train_model, y_train_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)
```

```python
params = {'n_estimators':[100,200,300],
    'max_depth':[6,8,10,12],
    'min_samples_leaf':[8,12,18],
    'min_samples_split':[8,16,20]}

grid_cv = GridSearchCV(model , param_grid=params , scoring='accuracy' , cv=5,
            verbose=1)

grid_cv.fit(X_train_model , y_train_model)
print('최적의 파라미터: ' , grid_cv.best_params_)
print('최고의 정확도: ' , grid_cv.best_score_)
```

```python
# 검증용 모델 생성 및 예측 후 점수 측정

best_dt_clf = grid_cv.best_estimator_
best_dt_clf.fit(X_train_model, y_train_model)


pred_val = best_dt_clf.predict_proba(X_train_val)[:,1]
print(roc_auc_score(y_train_val, pred_val))
```

```python
# 최종 제출 파일 생성

best_dt_clf.fit(X_train, y_train)
pred_final = best_dt_clf.predict_proba(X_test)[:,1]


final = pd.DataFrame({'CustomerId' : Id, 'Exited' : pred_final})
final.to_csv('final.csv', index=False)
```

```python
# 최종값 평가지표 출력
y_test = pd.read_csv('./test_label/y_test.csv')

print(roc_auc_score(y_test['Exited'] , pred_final))
```
